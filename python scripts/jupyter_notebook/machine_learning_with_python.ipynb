```python
from sklearn.tree import DecisionTreeClassifier #importing Desision tree classifier
from sklearn.model_selection import train_test_split #importing train test split function, to split the data into 2 groups, 1 for testing and 1 for training
from sklearn.metrics import accuracy_score # importing function to check the accuracy off the moddle with the previously mentioned testing data group 

x = df.drop(columns=['Drug']) #makes a new dataframe, without the drug column 
y = df['Drug'] # makes a new data frame with only the drug column 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3) # splits the data into a group to train the model and a group to test the model 

model = DecisionTreeClassifier() # setting the model 

model.fit(x_train, y_train) # training the model with the data 

predictions = model.predict(x_test) # letting the model predict the drug 

score = accuracy_score(y_test, predictions) # comparing the model to the actual values 

score # prints the accuracy score

```
